## Plot Nuscenes

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import pdb
plt.figure(figsize=(20,10))
for i in range(1,11):
  str_scene_num = str(i).zfill(2)
  b_firstline = True
  values = np.array([])
  scene_names = np.array([])
  with open('./nuScenes/' + str_scene_num + '/vel_ee_4_fde_most_likely_z.csv', newline='') as csvfile:
    fde_reader = csv.reader(csvfile)
    for row in fde_reader:
      if (b_firstline == True):
        b_firstline = False
        pass
      else:
        value = float(row[1])
        duration = float(row[4])
        num_nodes = float(row[5])
        scene_name = float(row[6])
        description = row[7]
        values = np.append(values,value)
        scene_names = np.append(scene_names,scene_name)
        # pdb.set_trace()

  scene_uniq_id = np.unique(scene_names)
  scene_val_sum = np.zeros_like(scene_uniq_id)
  scene_val_cnt = np.zeros_like(scene_uniq_id)
  for j in range(len(values)):
    scene_name = scene_names[j]
    idx = np.where(scene_uniq_id == scene_name)[0][0]
    scene_val_sum[idx] += values[j]
    scene_val_cnt[idx] += 1
  avg_val = scene_val_sum/scene_val_cnt
  pdb.set_trace()
  ## Plot
  ax=plt.subplot(2, 5, i)
  ax.bar(scene_uniq_id,avg_val)
  ax.set_title('FDE (m) avg for nuScene ' + str_scene_num)
  ax.set_xlabel('Scene Number')
  ax.set_ylabel('FDE (meters)')

plt.show()
```
